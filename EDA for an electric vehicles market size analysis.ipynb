### Exploratory Data Analysis For an Electric Vehicles (EV) Market Size Analysis|


To conduct Exploratory Data Analysis (EDA) for the Electric Vehicles Market Size Analysis, we will follow a systematic approach to understand the data, identify trends, and derive insights

#### 1. Data Cleaning and Preprocessing
- First, we will handle any missing values and ensure that data types are correct.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# Load the dataset
file_path = 'Electric_Vehicle_Population_Data.csv'
ev_data = pd.read_csv(file_path)

# Check for missing values
print(ev_data.isnull().sum())

# Handle missing values if necessary
# For simplicity, we will fill missing 'County' and 'City' with 'Unknown'
ev_data['County'].fillna('Unknown', inplace=True)
ev_data['City'].fillna('Unknown', inplace=True)

# Convert relevant columns to appropriate data types if necessary
ev_data['Postal Code'] = ev_data['Postal Code'].astype(str)
ev_data['Legislative District'] = ev_data['Legislative District'].astype(str)
ev_data['2020 Census Tract'] = ev_data['2020 Census Tract'].astype(str)

# Check data types
print(ev_data.dtypes)
```

#### 2. Exploratory Data Analysis
- a. Overview of the Data

```python
# Summary statistics
print(ev_data.describe())

# Unique values in categorical columns
print(ev_data['Make'].unique())
print(ev_data['Model'].unique())
print(ev_data['State'].unique())
```

- b. Registrations Over the Years

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=ev_data, x='Model Year', order=sorted(ev_data['Model Year'].unique()))
plt.title('Electric Vehicle Registrations by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

- c. Distribution of Electric Vehicle Types

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=ev_data, x='Electric Vehicle Type')
plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

- d. Geographical Distribution

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=ev_data, x='State', order=ev_data['State'].value_counts().index)
plt.title('Geographical Distribution of Electric Vehicles by State')
plt.xlabel('State')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=90)
plt.show()
```

- e. Popular Electric Vehicle Manufacturers

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=ev_data, x='Make', order=ev_data['Make'].value_counts().index)
plt.title('Distribution of Electric Vehicles by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=90)
plt.show()
```

- f. Electric Range Analysis

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=ev_data, x='Electric Range', bins=30, kde=True)
plt.title('Distribution of Electric Vehicle Ranges')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Frequency')
plt.show()
```

- g. Base MSRP Analysis

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=ev_data, x='Base MSRP', bins=30, kde=True)
plt.title('Distribution of Electric Vehicle MSRP')
plt.xlabel('Base MSRP ($)')
plt.ylabel('Frequency')
plt.show()
```

- Data visualization is essential; we must decide what charts to plot to better understand the data. In this article, we visualize our data using Matplotlib and Seaborn libraries.
  

- Matplotlib is a Python 2D plotting library used to draw basic charts we use Matplotlib.

- Univariate analysis can be done for both Categorical and Numerical variables.
  

- Categorical variables can be visualized using a Count plot, Bar Chart, Pie Plot, etc.
  

- Numerical Variables can be visualized using Histogram, Box Plot, Density Plot, etc.
  

- we have done a Univariate analysis using Histogram and  Box Plot for continuous Variables.
  

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numeric columns
num_cols = ev_data.select_dtypes(include='number').columns

# Iterate over numeric columns and plot histograms and boxplots
for col in num_cols:
    print(col)
    print('Skew :', round(ev_data[col].skew(), 2))
    plt.figure(figsize=(15, 4))
    
    plt.subplot(1, 2, 1)
    ev_data[col].hist(grid=False)
    plt.ylabel('count')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=ev_data[col])
    plt.show()
```
